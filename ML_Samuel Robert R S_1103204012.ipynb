```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
load = load_breast_cancer()
attributes = load['feature_names']
ds = pd.DataFrame(data=load['data'], columns=attributes)
ds = pd.DataFrame(np.c_[load['data'], load['target']], columns=np.append(load['feature_names'], ['target']))

ds.head()

sns.pairplot(ds, hue='target', vars=load['feature_names'][0:4])
```

```python
#x = nilai Freatures
#y = nilait target
X, y = load_breast_cancer(return_X_y=True)

print(f'Class = {set(y)}')
print(f'Baris data: {X.shape[0]}')
print(f'Kolom data: {X.shape[1]}')
```

```python
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)


#DECISION TREE
dec_tree = DecisionTreeClassifier(max_depth=3) 
dec_tree.fit(X=X_train, y = y_train)
y_pred = dec_tree.predict(X_test)
print("DECISON TREE")
print(classification_report(y_test, y_pred))


plt.rcParams['figure.dpi'] = 120
tree.plot_tree(dec_tree, fontsize=5)
plt.show()
```

```python
#RANDOM FOREST

rand_forest = RandomForestClassifier(n_estimators= 100, random_state=0)
#n_estimator banyak tree ayng digunakan

rand_forest.fit(X = X_train, y = y_train)

y_pred = rand_forest.predict(X_test)
print("RANDOM FOREST")
print(classification_report(y_test, y_pred))
```
